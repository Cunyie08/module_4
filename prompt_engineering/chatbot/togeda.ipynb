```python
from together import Together
```

```python
api_key = ""
```

```python
# initialize client
client = Together(api_key=api_key) # auth defaults to os.environ.get("TOGETHER_API_KEY")
```

| Parameter            | Purpose                     |
| -------------------- | --------------------------- |
| `messages`           | Conversation input          |
| `model`              | Which AI engine             |
| `max_tokens`         | Limit output length         |
| `stop`               | Force stopping              |
| `temperature`        | Creativity                  |
| `top_p`              | Nucleus sampling            |
| `top_k`              | Token filtering             |
| `min_p`              | Token probability threshold |
| `presence_penalty`   | New topic encouragement     |
| `frequency_penalty`  | Reduce repeated words       |
| `repetition_penalty` | Reduce token loops          |
| `logit_bias`         | Force/ban tokens            |
| `seed`               | Deterministic output        |
| `stream`             | Real-time output            |
| `logprobs`           | Return token probabilities  |
| `echo`               | Echo prompt                 |
| `n`                  | Number of completions       |
| `response_format`    | Force JSON or schema        |
| `tools`              | Enable tool calling         |
| `tool_choice`        | Control tool behavior       |
| `safety_model`       | Override safety settings    |


| Role          | Purpose                                       | Typical Usage                     |
| ------------- | --------------------------------------------- | --------------------------------- |
| **system**    | High-level instructions guiding the assistant | Set rules, tone, or persona       |
| **user**      | Message from the human                        | Questions, commands, conversation |
| **assistant** | Message from the AI                           | AI responses and reasoning        |
| **tool**      | Output returned by a called tool              | After tool execution              |
| **developer** | Developer instructions (optional mode)        | Structured or restricted tasks    |


```python
response = client.chat.completions.create(
    model="moonshotai/Kimi-K2-Thinking",
    messages=[
      {
        "role": "user",
        "content": "What are some fun things to do in New York?"
      }
    ]
)
print(response.choices[0].message.content)
```
