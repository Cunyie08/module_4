#### Zero-Shot Prompt  engineering

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()


prompt = "Translate the following English sentence into French: 'The weather is nice today'"

client = OpenAI(api_key=os.getenv("test_api"))
response = client.chat.completions.create(
    model="gpt-5.1",
    messages=[
        {"role": "developer", "content": "You are an assistive chatbot"},
        {"role": "user", "content": prompt}], temperature=0
        
)

print(response.choices[0].message.content)
```

#### One-Shot prompting

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()


prompt = """Detect the language of the following sentences:
"text: A plus tard"  -> language: "French
Now detect:
text: "Te amo!"
"""

client = OpenAI(api_key=os.getenv("test_api"))
response = client.chat.completions.create(
    model="gpt-5.1",
    messages=[
        {"role": "developer", "content": "You are an assistive chatbot"},
        {"role": "user", "content": prompt}], temperature=0
        
)

print(response.choices[0].message.content)
```
